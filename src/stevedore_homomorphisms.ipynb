```python
BSF=FreeGroup(2) 
```

```python
print(BSF) #x0=x0, x1=x2 in our presentation for the subgroup p_*(pi_1(unbranched cover))
```

```python
BSRelation=BSF([1,2,-1,-2,1,-2])
```

```python
BS=BSF.quotient([BSRelation]) #The group BS(1,2)
```

```python
a=BSF([2])*BSF([-1]) 
```

```python
print(a) #a in the subgroup <a^p,b>, a=x2x0^-1
```

```python
#Finds homomorphisms from BS(1,2) to S_p
#Prints images of x0, x2, and a
#want a to have order p
def find_BS_homs(p):    
    S_p=SymmetricGroup(p)
    gen_pairs=[]
    good_homs_p=[]
    for g_1 in S_p:
        for g_2 in S_p:
            pair=[g_1,g_2]
            hom_p=BSF.hom([S_p(g) for g in pair])
            if hom_p.image(BSRelation)==S_p.identity():
                good_homs_p.append(hom_p)
                print("x0->",hom_p.image(BSF([1])),"x2->",hom_p.image(BSF([2])),"a->",hom_p.image(a),"BSRelation->",hom_p.image(BSRelation))
    
```
