```python
# k >0 , t>=3 odd integer parameters
def S_k_t(k : int, t: int):
    S_k_t = Bridge_Trisection(k+2)
    S_k_t.red_tangle = Trivial_tangle(bridge_number=k+2,
                                               free_group=S_k_t.F,
                                               braid_word=singular_red_tangle_braid_crossings_list(k,t),
                                               strand_matching=singular_red_tangle_matching_list(k,t))
    S_k_t.blu_tangle = Trivial_tangle(bridge_number=k+2,
                                               free_group=S_k_t.F,
                                               braid_word=singular_blu_tangle_braid_crossings_list(k,t),
                                               strand_matching=singular_blu_tangle_matching_list(k,t))
    S_k_t.gre_tangle = Trivial_tangle(bridge_number=k+2, 
                                               free_group=S_k_t.F,
                                               braid_word=singular_gre_tangle_braid_crossings_list(k,t),
                                               strand_matching=singular_gre_tangle_matching_list(k,t))
    return S_k_t
```

```python
#Dihedral p-coloring
#p-colorable for t a multiple of p
def singular_surface_coloring(k,p):
    S = SymmetricGroup(p)
    #construct reflections over vertices 1 and 2 of p-gon
    pairs=[]
    for v in range(1,(p+1)/2):
        pairs.append((v+1,p-v+1))

    reflection_1=S('()')
    for pair in pairs:
        transposition=S(pair)
        reflection_1=reflection_1*transposition

    cycle_tuple=tuple([i+1 for i in range(p)])

    cycle=S(cycle_tuple)
    reflection_2=cycle^-1*reflection_1*cycle

    #For dihedral cover
    images_of_generators=[reflection_1,reflection_1^-1]+(k+1)*[reflection_2,reflection_2^-1] 
    return images_of_generators
```

```python
def colored_singular_surface(k,t,p):
    bridge_number = k+2
    
    sing_surf_F = FreeGroup(2*bridge_number)
    sing_surf_S = SymmetricGroup(p)
    sing_surf = S_k_t(k,t)
    sing_surf_tangles_dict = {
        'red': sing_surf.red_tangle,
        'blu': sing_surf.blu_tangle,
        'gre': sing_surf.gre_tangle,
    }
    
    
    
    sing_surf_images_of_generators=singular_surface_coloring(k,p)
    colored_sing_surf = Colored_bridge_trisection(sing_surf_F,
                                            sing_surf_S,
                                            sing_surf_tangles_dict,
                                            sing_surf_images_of_generators)
    return colored_sing_surf
```

```python
def display_invariants(k,t,p):
    col_sing_surf=colored_singular_surface(k,t,p)
    print("k=",k)
    print("t=",t)
    print("p=",p)
    print("Valid coloring on each tangle?")
    print(col_sing_surf.colored_tangles['red'].is_representation())
    print(col_sing_surf.colored_tangles['blu'].is_representation())
    print(col_sing_surf.colored_tangles['gre'].is_representation())
    print("fundamental group of branched cover")
    print(col_sing_surf.pi_1_branched_cover().simplified())
    print("trisection parameters")
    print(col_sing_surf.trisection_parameters_branched_cover())
    print("intersection form")
    print(col_sing_surf.intersection_form())
```
